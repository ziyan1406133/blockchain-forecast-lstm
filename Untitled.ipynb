```python
import numpy as np
import pandas as pd
import calendar
from math import sqrt
from datetime import date, timedelta
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,5)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM

gl_epoch = 20
lc_epoch = 5
neurons = 7
lag=7
```

### Data Collection

```python
df = pd.read_csv('blockchain_data/coin_EOS.csv')
df = df[df["Date"] <= "2020-12-31"]

def fix_date(text):
    text = text.split(" ")
    text = pd.Timestamp(text[0])
    return text

df['Date'] = df['Date'].apply(fix_date)
df
```

```python
plt.plot(df.Date, df.Close, label="Closing Price")

plt.title("Closing Price Pre-2021")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```

### Preprocessing

```python
def timeseries_to_supervised(df, n_in=1, n_out=1, dropnan=True):
    cols = list()
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
    for i in range(0, n_out):
        cols.append(df.shift(-i))
    agg = pd.concat(cols, axis=1)
    if dropnan:
        agg.dropna(inplace=True)
#     print(data)
#     print(agg)
    return agg.values
```

```python
def timeseries_to_supervised_v2(data, lag=1):
    columns = [data.shift(i) for i in range(1, lag+1)]
    columns.append(data)
    data = pd.concat(columns, axis=1)
    data.fillna(0, inplace=True)
    return data.values
```

```python
def scale(train, test):
    
    scaler = MinMaxScaler(feature_range=(-1, 1))
    scaler = scaler.fit(train)
    
    train = train.reshape(train.shape[0], train.shape[1])
    train_scaled = scaler.transform(train)
    
    test = test.reshape(test.shape[0], test.shape[1])
    test_scaled = scaler.transform(test)
    return scaler, train_scaled, test_scaled
```

```python
def invert_scale(scaler, X, value):
    new_row = [x for x in X] + [value]
    arr = np.array(new_row)
    arr = arr.reshape(1, len(arr))
    inverted = scaler.inverse_transform(arr)
    return inverted[0, -1]
```

```python
supervised_values = timeseries_to_supervised_v2(df["Close"], lag=lag)
supervised_values
```

```python
train_size = round(len(supervised_values) * 0.70)
test_size = len(supervised_values) - train_size
train, test = supervised_values[0:-test_size], supervised_values[-test_size:]
```

```python
both = list()
for t in train:
    both.append(t)
for t in test:
    both.append(t)

# scaler, train_scaled, test_scaled = scale(train, test)

scaler = MinMaxScaler(feature_range=(-1, 1))
scaler = scaler.fit(both)

train = train.reshape(train.shape[0], train.shape[1])
train_scaled = scaler.transform(train)

test = test.reshape(test.shape[0], test.shape[1])
test_scaled = scaler.transform(test)
scaled_values = list()

for t in train_scaled:
    scaled_values.append(t)
for t in test_scaled:
    scaled_values.append(t)
    
##train_scaled, test_scaled, 
scaled_values = np.array(scaled_values)
```

### Train and Predict

```python
def forecast_lstm(model, X):
    X = X.reshape(1, 1, len(X))
    yhat = model.predict(X, batch_size=1)
    return yhat[0,0]
```

```python
X, y = scaled_values[:, 0:-1], scaled_values[:, -1]
X = X.reshape(X.shape[0], 1, X.shape[1])
lstm_model = Sequential()
lstm_model.add(LSTM(neurons, batch_input_shape=(1, X.shape[1], X.shape[2]), stateful=True))
lstm_model.add(Dense(1))
lstm_model.compile(loss='mean_squared_error', optimizer='adam')

losses = list()
print("Training model...")
for i in range(gl_epoch):
    print("Epoch {}/{}".format(i+1, gl_epoch), end="")
    history = lstm_model.fit(X, y, epochs=lc_epoch, batch_size=1, verbose=False, shuffle=False)
    avg_loss = sum(history.history['loss']) / len(history.history['loss'])
    losses.append(avg_loss)
    print(", Average Loss : {}".format(avg_loss))
    lstm_model.reset_states()
```

```python
plt.plot(losses)

plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
predictions = list()
expected = list()
print("Predicting values...")
for i in range(len(scaled_values)):
    n = i + 1
    if ((n % 100 == 0 or n == len(scaled_values)) and i != 0):
        print("Predicted {} of {} data".format(n, len(scaled_values)))
    X, y = scaled_values[i, 0:-1], scaled_values[i, -1]
    output = forecast_lstm(lstm_model, X)
    output = invert_scale(scaler, X, output)
    y = invert_scale(scaler, X, y)
    predictions.append(output)
    expected.append(y)
#     print('Predicted={}, Expected={}'.format(output, y))
```

### Plot Prediction

```python
x_axis = [date_ for date_ in df.Date]

# print(len(x_axis))
# print(len(expected))
# print(len(predictions))

plt.plot(x_axis, expected, label='Expected')
plt.plot(x_axis, predictions, label='Predicted')

plt.title("Closing Price")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```

### Evaluation

```python
rmse = sqrt(mean_squared_error(expected, predictions))
print('Root Mean Squared Error: {:.3f}'.format(rmse))
```

### Forecasting Future Values

```python
number_of_forecast = 14
X = scaled_values[-1][-lag:]
forecast = list()
for n in range(number_of_forecast):
    output = forecast_lstm(lstm_model, X)
    X = np.append(X[1:], [output])
    output = invert_scale(scaler, X, output)
    forecast.append(output)
#     predictions
# forecast
```

```python
tail = number_of_forecast * 10

x_axis = []
y_forecast = []
expected_2 = []
predictions_2 = []
for i in range(tail, 0, -1):
    x_axis.append(df.Date.iloc[-i])
    y_forecast.append(None)
    expected_2.append(expected[-i])
    predictions_2.append(predictions[-i])

date_ = list()
for i in range(len(forecast)):
    new_date = x_axis[-1] + timedelta(days=1)
    date_.append(new_date)
    x_axis.append(new_date)
    y_forecast.append(forecast[i])
    expected_2.append(None)
    predictions_2.append(None)
    
df_ = pd.DataFrame(data={"Date": date_, "Value": forecast})
print("Forecast Result")
df_
```
